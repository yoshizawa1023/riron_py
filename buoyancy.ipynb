<h1>浮力シミュレーション</h1>
物体の落下運動の軌道を計算します。
<h2>パラメーター</h2>
<ul>
    <li>m:物体の質量</li>
    <li>g:重力加速度</li>
    <li>t:シミュレート時間</li>
    <li>fps:tあたりに演算する回数</li>
    <li>position:初期位置</li>
    <li>velocity:初速度</li>
<ul>

必要なライブラリをインポートします。

```python

import numpy as np
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import tqdm.notebook
```

パラメータを定義します。

```python
#定数 
m = 0.1       #質量
g = 9.8         #重力加速度
size = 0.1     #立方体の一片の長さ
water_level = 1 #水面の高さ
t = 1000
fps = 4

#変数
position = np.array([0,50])
velocity = np.array([10,0])
acceleration = np.array([0,0])

x = np.array([position[0]])
y = np.array([position[1]])
```

運動方程式を定義します。
F=maをaについて解いたものをし使います。

```python
#運動方程式
def equation_of_motion(f,m):
    return f/m
```

浮力の定義

```python
def buoyancy(water_level,p_y,size,G):
     if(p_y + size/2 < water_level):
          return size**3 * 1000 * G
     if ((water_level + size/2 - p_y) * size**2 * 1000 * G) > 0:
          return (water_level + size/2 - p_y) * size**2 * 1000 * G
     else:
          return 0
```

物体の位置,速度,加速度を更新します。

```python
#更新
def update():
    global position,velocity,acceleration

    f_0 = np.array([0,-m*g])
    f_1 = np.array([0,buoyancy(water_level=water_level,p_y=position[1],size=size,G=g)])
    f = f_0 + f_1
    print(f)
    acceleration = np.array([0,0])
    #重力
    acceleration = equation_of_motion(f,m)
    velocity = velocity + acceleration/fps
    position = position + velocity/fps
```

```python
for i in tqdm.tqdm(range(t*fps)):
    update()
    x = np.append(x,position[0])
    y = np.append(y,position[1])
plt.plot(x,y)
plot = plt.subplot()
plot.set_xlim([-10, 400])
plot.set_ylim([-25, 60])
plot.set_xlabel("x", size=14)
plot.set_ylabel("y", size=14)
plot.grid()
plot.set_aspect('equal')
plt.show()
```
