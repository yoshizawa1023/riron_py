<h1>落下運動シミュレーション</h1>
物体の落下運動の軌道を計算します。
<h2>パラメーター</h2>
<ul>
    <li>m:物体の質量</li>
    <li>g:重力加速度</li>
    <li>t:シミュレート時間</li>
    <li>fps:tあたりに演算する回数</li>
    <li>position:初期位置</li>
    <li>velocity:初速度</li>
<ul>

必要なライブラリをインポートします。

```python
import numpy as np
import matplotlib.pyplot as plt
import tqdm.notebook
```

パラメータを定義します。

```python
#定数 
m = 1   #質量
g = 9.8 #重力加速度
t = 10
fps = 60

#変数
position = np.array([0,0])
velocity = np.array([1,40])
acceleration = np.array([0,0])

x = np.array([position[0]])
y = np.array([position[1]])
```

運動方程式を定義します。
F=maをaについて解いたものをし使います。

```python
#運動方程式
def equation_of_motion(f,m):
    return f/m
```

物体の位置,速度,加速度を更新します。

```python
#更新
def update():
    global position,velocity,acceleration

    f = np.array([0,-m*g])

    acceleration = equation_of_motion(f,m)
    velocity = velocity + acceleration/fps
    position = position + velocity/fps
```

```python
for i in tqdm.tqdm(range(t*fps)):
    update()
    x = np.append(x,position[0])
    y = np.append(y,position[1])
plt.plot(x,y)

plt.show()
```
